# Protein Secondary Structure Prediction Using Python

## Overview

This project demonstrates a simple implementation of **protein secondary structure prediction** using the **Chou-Fasman statistical prediction method**.

The program:

* Reads protein sequences from a FASTA file
* Predicts secondary structure for each protein independently
* Assigns:

  * **H** = Alpha Helix
  * **E** = Beta Strand
  * **C** = Coil
* Produces secondary structure predictions for all sequences in the input file

---

## Prediction Method

### Chou-Fasman Method

The Chou-Fasman algorithm is one of the earliest computational approaches for protein secondary structure prediction.

The method is based on the observation that certain amino acids have different tendencies (**propensities**) to occur in:

* Alpha helices
* Beta strands
* Coil regions

For each amino acid:

* A helix propensity score is calculated
* A beta-strand propensity score is calculated

### Prediction Rules

```text
If Helix Score > Strand Score and Helix Score > 1.0
    → H (Alpha Helix)

If Strand Score > Helix Score and Strand Score > 1.0
    → E (Beta Strand)

Otherwise
    → C (Coil)
```

---

## Why Chou-Fasman?

The Chou-Fasman method was one of the first successful statistical approaches for predicting protein secondary structure directly from amino acid sequences.

Advantages:

* Easy to understand
* Computationally fast
* No external software required
* Suitable for educational purposes

````python

# ----------------------------------------------------- 
# Protein Secondary Structure Prediction
# Method: Simplified Chou-Fasman Algorithm 
# 
# H = Alpha Helix 
# E = Beta Strand 
# C = Coil 
# -----------------------------------------------------

# ----------------------------------------------------- 
# Read protein sequences from a FASTA file 
# Returns a dictionary:
# { 
# protein_id : sequence 
# } 
# -----------------------------------------------------

def read_fasta(filename):

    sequences = {}

    with open(filename, "r") as file:

        current_id = None
        current_seq = []

        for line in file:

            line = line.strip()

            if not line:
                continue

            if line.startswith(">"):

                if current_id:
                    sequences[current_id] = "".join(current_seq)

                current_id = line[1:]
                current_seq = []

            else:
                current_seq.append(line)

        # Save last sequence
        if current_id:
            sequences[current_id] = "".join(current_seq)

    return sequences
````

````python
# ----------------------------------------------------- 
# Chou-Fasman Alpha Helix Propensities 
# Values > 1 indicate preference for helices 
# -----------------------------------------------------

helix_propensity = {
    "A": 1.42, "C": 0.70, "D": 1.01, "E": 1.51,
    "F": 1.13, "G": 0.61, "H": 1.00, "I": 1.08,
    "K": 1.16, "L": 1.21, "M": 1.45, "N": 0.67,
    "P": 0.57, "Q": 1.11, "R": 0.98, "S": 0.77,
    "T": 0.83, "V": 1.06, "W": 1.08, "Y": 0.69
}

# ----------------------------------------------------- 
# Chou-Fasman Beta Strand Propensities 
# Values > 1 indicate preference for strands
# -----------------------------------------------------

strand_propensity = {
    "A": 0.83, "C": 1.19, "D": 0.54, "E": 0.37,
    "F": 1.38, "G": 0.75, "H": 0.87, "I": 1.60,
    "K": 0.74, "L": 1.30, "M": 1.05, "N": 0.89,
    "P": 0.55, "Q": 1.10, "R": 0.93, "S": 0.75,
    "T": 1.19, "V": 1.70, "W": 1.37, "Y": 1.40
}


         
````

````python
# ----------------------------------------------------- 
# Predict secondary structure
# 
# H = Alpha Helix 
# E = Beta Strand 
# C = Coil
# -----------------------------------------------------



def predict_secondary_structure(sequence):
    prediction = ""

    for aa in sequence:
        h_score = helix_propensity.get(aa, 1.0)
        e_score = strand_propensity.get(aa, 1.0)

        if h_score > e_score and h_score > 1.0:
            prediction += "H"
        elif e_score > h_score and e_score > 1.0:
            prediction += "E"
        else:
            prediction += "C"

    return prediction
````

````python
# Read proteins from FASTA
proteins = read_fasta('dao.fa')


# ----------------------------------------------------- 
# Predict secondary structure for each protein 
# -----------------------------------------------------

for protein_id, protein_sequence in proteins.items():

    predicted_structure = predict_secondary_structure(
        protein_sequence
    )

    print("=" * 60)
    print("Protein ID:", protein_id)

    print("\nProtein Sequence:")
    print(protein_sequence)

    print("\nPredicted Secondary Structure:")
    print(predicted_structure)
````
